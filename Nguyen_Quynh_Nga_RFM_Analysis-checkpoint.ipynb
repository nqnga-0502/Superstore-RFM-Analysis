```python
rtn = pd.read_excel('Dataset.xlsx',sheet_name='Return')
rtn.columns=[col.lower().replace(' ','_') for col in rtn.columns]
rtn.info()
```

```python
ord_rtn = group_order.merge(rtn,how='left',on='order_id')
ord_rtn.head(15)
```

```python
valid=ord_rtn[ord_rtn['returned']!='Yes'].drop('returned',axis=1)
valid.head(10)
valid
```

```python
rfm = valid.groupby('customer_id').agg(max_date= ('order_date','max'),frequency=('order_id','count'),monetary=('sales','sum')).reset_index()

rfm
```

```python
last_date=dt.datetime.strptime('2017-12-31','%Y-%m-%d')
```

```python
rfm['recency']=(last_date-rfm['max_date']).dt.days
rfm=rfm.drop('max_date',axis=1)

rfm
```

```python
rfm=rfm[['customer_id','recency','frequency','monetary']]
rfm
```

```python
rfm['r_score'] = pd.qcut(rfm['recency'], 5, ['5','4','3','2','1'])
rfm['f_score'] = pd.qcut(rfm['frequency'], 5, ['1','2','3','4','5'])
rfm['m_score'] = pd.qcut(rfm['monetary'], 5, ['1','2','3','4','5'])
rfm
```

```python
rfm['rfm_score']=(rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)).astype(int)
rfm.head()
```

```python
segment=pd.read_excel('Dataset.xlsx',sheet_name = 'Segmentation')
segment.columns = [col.lower().replace(' ','_') for col in segment.columns]
segment
```

```python
segment['rfm_score']=segment['rfm_score'].str.split(',')
segment=segment.explode('rfm_score').reset_index(drop=True)
segment['rfm_score']=segment['rfm_score'].astype(int)
segment.head(10)
```

```python
rfm_seg = rfm.merge(segment, how='left', on='rfm_score')
rfm_seg
```

```python
palette_1=sns.color_palette('GnBu_r',11) 
palette_2=sns.color_palette('coolwarm_r',11)
palette_3=sns.color_palette('RdBu_r',11)
palette_4=sns.color_palette('RdBu',11)
palette_5=sns.color_palette('crest_r',11)
```

```python
col_name=['recency','frequency','monetary']
sns.color_palette('RdBu')
sns.set_theme(style="ticks", font_scale=0.8)

for col in col_name:
    dis_plot=sns.displot(rfm_seg[col], kde=True, stat='density', height=3.5, aspect = 2)
    dis_plot.fig.suptitle('Distribution of %s' % col, y=1.1,fontsize=12, fontweight='bold')

    
    plt.show()
```

```python
grp = rfm_seg.groupby('segment').agg(cust_count=('customer_id','count'),total_mon=('monetary','sum'),total_freq=('frequency','sum'),avg_rec=('recency','mean'),avg_freq=('frequency',np.average)).reset_index()

grp
```

```python
grp['perc_cust']= grp['cust_count']/grp['cust_count'].sum()
grp['perc_mon'] = grp['total_mon']/grp['total_mon'].sum()
grp['perc_feq']=grp['total_freq']/grp['total_freq'].sum()
grp
```

```python
r_chart = sns.barplot(y='segment',x='cust_count',data=grp.sort_values('cust_count',ascending=False), palette = palette_1)
r_chart.set_title('Total customer by RFM Segments',fontsize=12, fontweight='bold')
r_chart.set(xlabel='Segment',ylabel='Average Recency')
r_chart.spines['top'].set_visible(False)
r_chart.spines['right'].set_visible(False)

for i, v in enumerate(grp.sort_values('cust_count',ascending=False)['cust_count'].round(2)):
    plt.text(v+1 , i+0.2 , str(v), color='black', fontsize=9)
plt.show()
```

```python
m_chart = sns.barplot(x='segment',y='total_mon',data=grp.sort_values('total_mon',ascending=False), palette = palette_3)
m_chart.set(ylabel='Revenue',xlabel=None)
plt.xticks(rotation=90)
plt.legend(['Revenue'],loc='upper left', fontsize=10)
m_chart.spines['top'].set_visible(False)
labels = grp.sort_values('total_mon',ascending=False)['total_mon'].round(2)


f_chart= m_chart.twinx()
sns.lineplot(x='segment',y='avg_freq',data=grp.sort_values('total_mon',ascending=False),color='lightcoral', marker='o', ax=f_chart)
f_chart.set(ylabel='Average number of orders')
plt.legend(['Average number of orders'],loc='upper right',fontsize=10)
f_chart.spines['top'].set_visible(False)
f_chart.set_title('Average Frequency vs Total Revenue by RFM Segments', fontsize=12, fontweight='bold')


plt.show()
```
